```python
import random

# Player choice counters
r = 0 # rock
s = 0 # scissors
p = 0 # paper

# Game counter
count = 0
# Win counter (for player)
wins = 0

#Winning end-phrases
phrases = ('Paper covers Rock, so Paper beats Rock.',
           'Rock smashes Scissors, so Rock beats Scissors.',
           'Scissors can cut Paper, so Scissors beats Paper.')

# Choices
options = ('rock','scissors','paper')

############################################################################################################################
# Unknown number of rounds (play until user quits)
play = True
while play == True:
    
    #Beginning of each round
    
    # Update players previous choices in one variable
    player_data = (r,s,p)
    
    # Computer makes choice
    # Completely random choice for first game
    if count == 0:
        comp_choice = options[random.randint(0,2)]
    
    # Base choice on players previous moves
    else:
        # Guessing which option player will pick next
        # player_data = [# of rock, # of scissors, # of paper]
        # Best guess for what player will choose next
        guess = max(player_data)
        best_guess = []
        for i, x in enumerate(player_data):
            if x == guess:
                best_guess.append(i)        ###I THINK THERE IS A BETTER WAY TO DO THIS
        
        if len(best_guess) == 1:
            # String of the guess of players next choice
            final_guess = options[best_guess[0]]
            # Computers next choice
            if final_guess == 'rock':
                comp_choice = 'paper'
            elif final_guess is 'scissors':
                comp_choice = 'rock'
            elif final_guess is 'paper':
                comp_choice = 'scissors'
        
        elif len(best_guess) == 2:
            # Player is equally likely to choose one of two options
            
            # Best course of action is to have computer choose the winning position b/w the two
            # This means worst-case is draw, best-case is win for computer assuming prediction
            # method for player is accurate
            
            # Rock and Scissors
            if best_guess[0] == 0 and best_guess[1] == 1:
                comp_choice = options[0]
            # Rock and Paper
            elif best_guess[0] == 0 and best_guess[1] == 2:
                comp_choice = options[2]
            #Scissors and Paper
            elif best_guess[0] == 1 and best_guess[1] == 2:
                comp_choice = options[1]
            
        else:
            # All three optiosn equally likey, thus computer's choice is arbitrary
            comp_choice = options[random.randint(0,2)]
    
    # Player makes choice
    while True:
        player = input('Rock, Paper, Scissors, or Quit?\n')
        player = player.lower()
        if player == 'quit':
            play = False
            break
        elif player == 'rock':
            r += 1
            break
        elif player == 'paper':
            p += 1
            break
        elif player == 'scissors':
            s += 1
            break
        else:
            print('Invalid Input! Try Again:')
    
############################################################################################################################    
    # Results (if player wants to play)
    if play == True:
        count += 1
        
        # Determine winner
        
        # Draws
        if comp_choice == player:
            print("It's a draw!")
        
        # With winner
        elif comp_choice == 'rock':
            if player == 'paper':
                print(phrases[0])
                wins += 1
            elif player == 'scissors':
                print(phrases[1])
                
        elif comp_choice == 'scissors':
            if player == 'paper':
                print(phrases[2])
            elif player == 'rock':
                print(phrases[1])
                wins += 1
                
        elif comp_choice == 'paper':
            if player == 'rock':
                print(phrases[0])
            elif player == 'scissors':
                print(phrases[2])
                wins += 1
    
    # Exit phrase if player quits ADD PLAYER DATA TO OUTPUT
    else:
        print('You played ' + str(count) + ' games, and you won ' + str(wins) + ' games\n')
```
